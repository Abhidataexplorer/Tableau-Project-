```python
different_columns_for_boxplot = ['population_country', 'tax_revenue_country_country', 'gross_tertiary_education_enrollment']

# Create boxplot for different columns
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[different_columns_for_boxplot])
plt.title('Boxplot of Different Columns')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.xlabel('Columns')
plt.ylabel('Values')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.scatter(data['age'], data['finalWorth'], alpha=0.5)  # alpha controls the transparency of the points
plt.title('Scatter Plot of Age vs Final Worth')
plt.xlabel('Age')
plt.ylabel('Final Worth')
plt.grid(True)  # Add grid lines
plt.show()
```

```python
sns.pairplot(data[['age', 'finalWorth', 'rank', 'tax_revenue_country_country', 'total_tax_rate_country']])
plt.show()
```

```python
correlation_matrix = data[['age', 'finalWorth', 'rank','birthDate','birthYear','birthMonth']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
